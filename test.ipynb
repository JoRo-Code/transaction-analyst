```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
wgr_df = pd.read_csv('data/wgr.csv', delimiter='\t', encoding='utf-16')
qliro_df = pd.read_csv('data/qliro.csv', sep=';')
```

```python
wgr_df = wgr_df[wgr_df['Payment method'] == 'QLIROCHECKOUT']
```

```python

wgr_df = wgr_df[['Order ID', 'Total amount excl. VAT', 'Total VAT', 'Order time']]








```

```python
wgr_df
```

```python
# Remove 'WGR' prefix from Order ID column
qliro_df['Butiksordernummer'] = qliro_df['Butiksordernummer'].str.replace('WGR', '')


# Select relevant columns
qliro_df = qliro_df[['Butiksordernummer', 'Betalning transaktionsreferens', 'Belopp','Avräkningsstatus', 'Avräkningsdatum', 'Transaktionsslutdatum']]

```

```python
qliro_df.head()
```

```python
wgr_df['Order ID'] = wgr_df['Order ID'].astype(str)
qliro_df['Butiksordernummer'] = qliro_df['Butiksordernummer'].astype(str)

merged_df = pd.merge(
    wgr_df,
    qliro_df, 
    left_on='Order ID',
    right_on='Butiksordernummer',
    how='inner'
)
```

```python
merged_df
```

```python
# Define the period
start_date = '2024-11-01'
end_date = '2024-11-30'

# Convert datetime columns if not already
merged_df['Order time'] = pd.to_datetime(merged_df['Order time'])
merged_df['Avräkningsdatum'] = pd.to_datetime(merged_df['Avräkningsdatum'])

# Split all orders into settled, late settled and unsettled orders
settled_orders = merged_df[
    (merged_df['Avräkningsstatus'] == 'Settled') &
    (merged_df['Avräkningsdatum'].dt.strftime('%Y-%m-%d') <= end_date)
]

late_settled_orders = merged_df[
    (merged_df['Avräkningsstatus'] == 'Settled') &
    (merged_df['Avräkningsdatum'].dt.strftime('%Y-%m-%d') > end_date)
]

unsettled_orders = merged_df[
    merged_df['Avräkningsstatus'] != 'Settled'
]

# Further split each group into period and non-period orders
period_settled = settled_orders[
    (settled_orders['Order time'].dt.strftime('%Y-%m-%d') >= start_date) & 
    (settled_orders['Order time'].dt.strftime('%Y-%m-%d') <= end_date)
]

non_period_settled = settled_orders[
    (settled_orders['Order time'].dt.strftime('%Y-%m-%d') < start_date) | 
    (settled_orders['Order time'].dt.strftime('%Y-%m-%d') > end_date)
]

period_late_settled = late_settled_orders[
    (late_settled_orders['Order time'].dt.strftime('%Y-%m-%d') >= start_date) & 
    (late_settled_orders['Order time'].dt.strftime('%Y-%m-%d') <= end_date)
]

non_period_late_settled = late_settled_orders[
    (late_settled_orders['Order time'].dt.strftime('%Y-%m-%d') < start_date) | 
    (late_settled_orders['Order time'].dt.strftime('%Y-%m-%d') > end_date)
]

period_unsettled = unsettled_orders[
    (unsettled_orders['Order time'].dt.strftime('%Y-%m-%d') >= start_date) & 
    (unsettled_orders['Order time'].dt.strftime('%Y-%m-%d') <= end_date)
]

non_period_unsettled = unsettled_orders[
    (unsettled_orders['Order time'].dt.strftime('%Y-%m-%d') < start_date) | 
    (unsettled_orders['Order time'].dt.strftime('%Y-%m-%d') > end_date)
]
```
